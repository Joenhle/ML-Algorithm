```python
import pandas as pd
import numpy as np
import math

def get_data():
    file = pd.read_csv(open("/Users/huangjunhao/大学/大三/大三上/机器学习/实践题/实验一放假预测/housing.csv"))#将csv文件读取为pd格式

    file = file.dropna(axis = 0,how = 'any')#清洗nan
    data = file.to_numpy()#转换为np
    list_of_ocean = list(file['ocean_proximity'].unique())#将字符串列转化为数字
    for line in data:
        line[-1] = list_of_ocean.index(line[-1])+1
    #切片出特征和结果
    y = data[:,-2]
    x = np.delete(data, -2, axis=1)
    x = np.c_[[1]*len(x),x]
    return x,y
```

```python
#最小二乘法
def least_square_method(x,y): 
    # w = ((x^T*x)^(-1))*x^T*y
    w = np.dot(np.dot(np.matrix(np.array(np.dot(x.T,x),'float32')).I,x.T),y)
    return w

x,y = get_data()
w = least_square_method(x,y)
print(w)
print(np.dot(w,x[3]))
```

```python
# 批量梯度下降
def batch_gradient_descnt(x,y):
    step = 0.00001
    max_iteration = 100
    w = [1]*len(x[0])
    N = len(x)
    
    for i in range(max_iteration):
        print(w)
        last_w = []
        
        for j in range(len(w)):
            sum = 0;
            for k in range(len(x)):
                sum += (np.dot(w,x[k])-y[k])*x[k][j]
            last_w.append(w[j] - step*(sum/N)) 
        w = last_w.copy()
    return w

w = batch_gradient_descnt(x,y)
print(w)

    
```
